# Mean Reversion Estrategy 1

### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import statsmodels.stats.multitest as mt
import arch.bootstrap as boot
import pyalgotrade.barfeed.csvfeed as csvfeed
import pyalgotrade.bar as bar
import pyalgotrade.strategy as strategy
import pyalgotrade.technical.bollinger as bollinger
import pyalgotrade.broker as broker
import pyalgotrade.stratanalyzer.returns as ret
import pyalgotrade.plotter as plotter
import datetime as dt
import itertools
import time
```

### Create Strategy Class

```python
class MeanStrategy1(strategy.BacktestingStrategy):

    # 2.1. Define Strategy Initialization Function
    def __init__(self, feed, instrument, nper, nstdev):
        super(MeanStrategy1, self).__init__(feed, 10000)
        self.position = None
        self.instrument = instrument
        self.setUseAdjustedValues(True)
        self.bbands = bollinger.BollingerBands(feed[instrument].getPriceDataSeries(), nper, nstdev)

    # 2.2. Define Get Technical Indicators Functions
    def getBBands(self):
        return self.bbands

    # 2.3. Define Overriding Strategy Functions

    # onEnterCanceled: Get notified when order submitted to enter a position was canceled and update position
    def onEnterCanceled(self, position):
        self.position = None

    # onExitOk: Get notified when order submitted to exit a position was filled and update position
    def onExitOk(self, position):
        self.position = None

    # onExitCanceled: Get notified when order submitted to exit a position was canceled and re-submit order
    def onExitCanceled(self, position):
         self.position.exitMarket()

    # 2.4. Define Trading Strategy Function

    # Mean-Reversion Strategy
    # Enter Long Order = Buy when Close < Lower Band
    # Exit Order = Sell when Close > Upper Band

    def onBars(self, bars):
        if self.bbands.getLowerBand()[-1] is None:
            return

        if self.position is None:
            if bars[self.instrument].getPrice() < self.bbands.getLowerBand()[-1]:
                # 95% equity investment for difference between order day Close price and next day Open price
                # number of shares can also be a fixed quantity for all transactions (ex. self.shares = 10)
                self.shares = int(self.getBroker().getCash() * 0.95 / bars[self.instrument].getPrice())
                self.position = self.enterLong(self.instrument, self.shares, goodTillCanceled=True)

        elif bars[self.instrument].getPrice() > self.bbands.getUpperBand()[-1] and not self.position.exitActive():
            self.position.exitMarket()
```

# Define Run Strategy Function

```python
def MeanStrategyRun1(nper, nstdev, chart):

    # 3.1. Create Instruments object with stock tickers
    instruments = ['SPY']

    # 3.2. Load CSV Feed previously downloaded or read
    feed = csvfeed.GenericBarFeed(bar.Frequency.DAY)
    feed.addBarsFromCSV(instruments[0], './Advanced-Trading-Analysis-Data.txt', skipMalformedBars=True)

    # 3.3. Evaluate Strategy with CSV Feed and Technical Indicator Parameters
    meanStrategy1 = MeanStrategy1(feed, instruments[0], nper, nstdev)

    # 3.4. Set Strategy Commission
    meanStrategy1.getBroker().setCommission(broker.backtesting.FixedPerTrade(6))

    # 3.5. Attach Strategy Trading Statistics Analyzers
    retAnalyzer = ret.Returns(maxLen=2518)
    meanStrategy1.attachAnalyzer(retAnalyzer)

    # 3.6. Attach Strategy Plotter
    plt = plotter.StrategyPlotter(meanStrategy1, plotPortfolio=False)
    plt.getInstrumentSubplot('SPY').addDataSeries('Lower', meanStrategy1.getBBands().getLowerBand())
    plt.getInstrumentSubplot('SPY').addDataSeries('Middle', meanStrategy1.getBBands().getMiddleBand())
    plt.getInstrumentSubplot('SPY').addDataSeries('Upper', meanStrategy1.getBBands().getUpperBand())

    # 3.7. Run Strategy
    meanStrategy1.run()

    # 3.8. Calculate Strategy Returns
    datesReturns = retAnalyzer.getReturns().getDateTimes()[:]
    dailyReturns = retAnalyzer.getReturns()[:]
    dailyReturns = pd.DataFrame(dailyReturns).set_index(pd.DatetimeIndex(datesReturns))

    # 3.9. Plot Strategy
    if chart == True:
        plt.plot(fromDateTime=dt.datetime(2016, 1, 1), toDateTime=dt.datetime(2016, 12, 31))

    return dailyReturns
```

```python
# 3.10. Plot Strategy Example
MeanStrategyRun1(20, 2, True)
```

### Do Strategy Parameters Optimization and Calculate Performance Metrics

```python
# 4.1. Create Strategy Optimization Parameters Combinations
nper = (20, 25)
nstdev = (1, 2)
pool = [nper, nstdev]
```

```python
# 4.2. Calculate Benchmark Daily Returns
# 100% equity investment for buy and hold benchmark
data = pd.read_csv('./Advanced-Trading-Analysis-Data.txt', index_col='Date Time', parse_dates=True)
mean1DailyReturns = data['Adj Close'].pct_change(1)
mean1DailyReturns[0] = 0
mean1DailyReturns = pd.DataFrame(mean1DailyReturns)
```

```python
# 4.3. Do Strategy Optimization
mean1StartOptimization = time.time()
print('')
print('== Strategy Parameters Optimization ==')
print('')
print('Parameters Combinations (nper, nstdev):')
for n in itertools.product(*pool):
    print(n)
    mean1DailyReturns.insert(len(mean1DailyReturns.columns), n, MeanStrategyRun1(n[0], n[1], False))
mean1EndOptimization = time.time()
mean1DailyReturns.columns = ['B&H', 'Mn1Ret1', 'Mn1Ret2', 'Mn1Ret3', 'Mn1Ret4']
print('')
print('Optimization Running Time: ', round(mean1EndOptimization - mean1StartOptimization, 4), ' seconds')
print('')
```

```python
# 4.4. Define Cumulative Returns and Performance Metrics Functions
def CumulativeReturns(dailyReturns):
    cumulativeReturns = np.cumprod(dailyReturns + 1) ** (252 / len(dailyReturns)) - 1
    return cumulativeReturns

def PerformanceMetrics(dailyReturns):
    annualizedReturn = (np.cumprod(dailyReturns + 1) ** (252 / len(dailyReturns)) - 1)[-1]
    annualizedStdDev = np.std(dailyReturns) * np.sqrt(252)
    annualizedSharpe = annualizedReturn / annualizedStdDev
    return annualizedReturn, annualizedStdDev, annualizedSharpe
```

```python
# 4.5. Chart Cumulative Returns Comparison
mean1CumulativeReturns = mean1DailyReturns.apply(CumulativeReturns, axis=0)
plt.plot(mean1CumulativeReturns['B&H'], label='B&H')
plt.plot(mean1CumulativeReturns['Mn1Ret1'], label='Mn1Ret1')
plt.plot(mean1CumulativeReturns['Mn1Ret2'], label='Mn1Ret2')
plt.plot(mean1CumulativeReturns['Mn1Ret3'], label='Mn1Ret3')
plt.plot(mean1CumulativeReturns['Mn1Ret4'], label='Mn1Ret4')
plt.title('Strategy Parameters Optimization Cumulative Returns')
plt.legend(loc='upper left')
plt.show()
```

```python
# 4.6. Calculate Performance Metrics and Print Summary Table
mean1PerformanceMetrics = mean1DailyReturns.apply(PerformanceMetrics, axis=0)
mean1PerformanceSummary = [{'0': 'Annualized:', '1': 'B&H', '2': 'Mn1Ret1', '3': 'Mn1Ret2', '4': 'Mn1Ret3',
                             '5': 'Mn1Ret4'},
        {'0': 'Return',
         '1': np.round(mean1PerformanceMetrics[0][0], 4), '2': np.round(mean1PerformanceMetrics[1][0], 4),
         '3': np.round(mean1PerformanceMetrics[2][0], 4), '4': np.round(mean1PerformanceMetrics[3][0], 4),
         '5': np.round(mean1PerformanceMetrics[4][0], 4)},
        {'0': 'Standard Deviation',
         '1': np.round(mean1PerformanceMetrics[0][1], 4), '2': np.round(mean1PerformanceMetrics[1][1], 4),
         '3': np.round(mean1PerformanceMetrics[2][1], 4), '4': np.round(mean1PerformanceMetrics[3][1], 4),
         '5': np.round(mean1PerformanceMetrics[4][1], 4)},
        {'0': 'Sharpe Ratio (Rf=0%)',
         '1': np.round(mean1PerformanceMetrics[0][2], 4), '2': np.round(mean1PerformanceMetrics[1][2], 4),
         '3': np.round(mean1PerformanceMetrics[2][2], 4), '4': np.round(mean1PerformanceMetrics[3][2], 4),
         '5': np.round(mean1PerformanceMetrics[4][2], 4)}]
mean1PerformanceSummary = pd.DataFrame(mean1PerformanceSummary)
print('')
print('== Strategy Parameters Optimization Performace Summary ==')
print('')
print(mean1PerformanceSummary)
print('')
```

# Do Multiple Hypothesis Testing P-Values Adjustments

```python
# 5.1. Calculate Multiple Hypothesis Testing P-Values
mean1MultipleTTests = mean1DailyReturns.iloc[:, 1:5].apply(st.ttest_1samp, axis=0, popmean=0, nan_policy='omit')
mean1MultiplePValues = (mean1MultipleTTests[0][1], mean1MultipleTTests[1][1], mean1MultipleTTests[2][1],
                         mean1MultipleTTests[3][1])

# 5.2. Adjust Multiple Hypothesis Testing P-Values Calculations
mean1MultiplePValuesFWE = mt.multipletests(mean1MultiplePValues, alpha=0.05, method='bonferroni',
                                                   is_sorted=False, returnsorted=False)
mean1MultiplePValuesFDR = mt.multipletests(mean1MultiplePValues, alpha=0.05, method='fdr_bh',
                                                   is_sorted=False, returnsorted=False)

# 5.3. Print Multiple Hypothesis Testing P-Values Adjustment Summary Table
mean1MultiplePValuesSummary = [{'0': '', '1': 'Mn1Ret1', '2': 'Mn1Ret2', '3': 'Mn1Ret3', '4': 'Mn1Ret4'},
        {'0': 'PValues',
         '1': np.round(mean1MultiplePValues[0], 4), '2': np.round(mean1MultiplePValues[1], 4),
         '3': np.round(mean1MultiplePValues[2], 4), '4': np.round(mean1MultiplePValues[3], 4)},
        {'0': 'PValues FWE',
         '1': np.round(mean1MultiplePValuesFWE[1][0], 4), '2': np.round(mean1MultiplePValuesFWE[1][1], 4),
         '3': np.round(mean1MultiplePValuesFWE[1][2], 4), '4': np.round(mean1MultiplePValuesFWE[1][3], 4)},
        {'0': 'PValues FDR',
         '1': np.round(mean1MultiplePValuesFDR[1][0], 4), '2': np.round(mean1MultiplePValuesFDR[1][1], 4),
         '3': np.round(mean1MultiplePValuesFDR[1][2], 4), '4': np.round(mean1MultiplePValuesFDR[1][3], 4)}]
mean1MultiplePValuesSummary = pd.DataFrame(mean1MultiplePValuesSummary)
print('')
print('== Multiple Hypothesis Testing P-Values Adjustments ==')
print('')
print(mean1MultiplePValuesSummary)
print('')
```

### Do Individual Time Series Bootstrap P-Value Multiple Comparison Adjustment

```python
# 6.1. Define Bootstrap Mean Function
def bmean(x):
    return x.mean(0)

# 6.2. Do Individual Time Series Bootstrap
mean1StartBoot = time.time()
print('')
print('== Individual Time Series Bootstrap ==')
print('')
mean1Boot = boot.CircularBlockBootstrap(block_size=10, x=mean1DailyReturns.iloc[:, 4])
mean1BootMeans = mean1Boot.apply(func=bmean, reps=1000)
mean1BootIntervals = mean1Boot.conf_int(func=bmean, reps=1000, method='percentile', size=0.95, tail='two')
mean1EndBoot = time.time()
print('')
print('Bootstrap Running Time: ', round(mean1EndBoot - mean1StartBoot, 4), ' seconds')
print('')
```

```python
# 6.3. Chart Individual Time Series Bootstrap Histogram
plt.hist(mean1BootMeans, bins=20, density=True, label='BootMeans')
plt.title('Population Mean Probability Distribution Simulation')
plt.axvline(x=np.mean(mean1DailyReturns.iloc[:, 4]), color='purple', linestyle='--', label='mean(Mn1Ret4)')
plt.axvline(x=np.mean(mean1BootMeans), color='red', linestyle='--', label='mean(BootMeans)')
plt.axvline(x=0, color='orange', linestyle='--')
plt.axvline(x=mean1BootIntervals[0], color='green', linestyle='--', label='BootLowerCI')
plt.axvline(x=mean1BootIntervals[1], color='green', linestyle='--', label='BootUpperCI')
plt.ylabel('Density')
plt.xlabel('Bin Edges')
plt.legend(loc='upper right')
plt.show()
```

```python
# 6.4. Calculate Individual Time Series Bootstrap P-Value
mean1BootPValue = 2 * min(np.mean(mean1BootMeans <= 0), np.mean(mean1BootMeans > 0))
```

```python
# 6.5. Adjust Individual Time Series Bootstrap P-Value Calculation
mean1BootPValueFWE = 1 - (1 - mean1BootPValue) ** 4
print('')
print('== Individual Time Series Bootstrap Hypothesis Testing ==')
print('')
print('Mn1Ret4 P-Value:', np.round(mean1BootPValue, 4))
print('Mn1Ret4 P-Value FWE:', np.round(mean1BootPValueFWE, 4))
```
