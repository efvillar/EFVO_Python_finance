# Mean Reversion Strategy

### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import statsmodels.stats.multitest as mt
import arch.bootstrap as boot
import pyalgotrade.barfeed.csvfeed as csvfeed
import pyalgotrade.bar as bar
import pyalgotrade.strategy as strategy
import pyalgotrade.technical.rsi as rsi
import pyalgotrade.broker as broker
import pyalgotrade.stratanalyzer.returns as ret
import pyalgotrade.plotter as plotter
import datetime as dt
import itertools
import time
```

### Create Strategy Class

```python
class MeanStrategy2(strategy.BacktestingStrategy):

    # 2.1. Define Strategy Initialization Function
    def __init__(self, feed, instrument, nper, lowerThreshold, upperThreshold):
        super(MeanStrategy2, self).__init__(feed, 10000)
        self.position = None
        self.instrument = instrument
        self.setUseAdjustedValues(True)
        self.rsi = rsi.RSI(feed[instrument].getPriceDataSeries(), nper)
        self.lowerThreshold = lowerThreshold
        self.upperThreshold = upperThreshold

    # 2.2. Define Get Technical Indicators Functions
    def getRSI(self):
        return self.rsi

    # 2.3. Define Overriding Strategy Functions

    # onEnterCanceled: Get notified when order submitted to enter a position was canceled and update position
    def onEnterCanceled(self, position):
        self.position = None

    # onExitOk: Get notified when order submitted to exit a position was filled and update position
    def onExitOk(self, position):
        self.position = None

    # onExitCanceled: Get notified when order submitted to exit a position was canceled and re-submit order
    def onExitCanceled(self, position):
         self.position.exitMarket()

    # 2.4. Define Trading Strategy Function

    # Mean-Reversion Relative-Strength Strategy
    # Enter Long Order = Buy when RSI < Lower Threshold
    # Exit Order = Sell when RSI > Upper Threshold

    def onBars(self, bars):
        if self.rsi[-1] is None:
            return

        if self.position is None:
            if self.rsi[-1] < self.lowerThreshold:
                # 95% equity investment for difference between order day Close price and next day Open price
                # number of shares can also be a fixed quantity for all transactions (ex. self.shares = 10)
                self.shares = int(self.getBroker().getCash() * 0.95 / bars[self.instrument].getPrice())
                self.position = self.enterLong(self.instrument, self.shares, goodTillCanceled=True)

        elif self.rsi[-1] > self.upperThreshold and not self.position.exitActive():
            self.position.exitMarket()
```

### Define Run Strategy Function

```python
def MeanStrategyRun2(nper, lowerThreshold, upperThreshold, chart):

    # 3.1. Create Instruments object with stock tickers
    instruments = ['SPY']

    # 3.2. Load CSV Feed previously downloaded or read
    feed = csvfeed.GenericBarFeed(bar.Frequency.DAY)
    feed.addBarsFromCSV(instruments[0], './Advanced-Trading-Analysis-Data.txt', skipMalformedBars=True)

    # 3.3. Evaluate Strategy with CSV Feed and Technical Indicator Parameters
    meanStrategy2 = MeanStrategy2(feed, instruments[0], nper, lowerThreshold, upperThreshold)

    # 3.4. Set Strategy Commission
    meanStrategy2.getBroker().setCommission(broker.backtesting.FixedPerTrade(6))

    # 3.5. Attach Strategy Trading Statistics Analyzers
    retAnalyzer = ret.Returns(maxLen=2518)
    meanStrategy2.attachAnalyzer(retAnalyzer)

    # 3.6. Attach Strategy Plotter
    plt = plotter.StrategyPlotter(meanStrategy2, plotPortfolio=False)
    plt.getOrCreateSubplot('RSI').addDataSeries('RSI', meanStrategy2.getRSI())
    plt.getOrCreateSubplot('RSI').addLine('Lower', level=30)
    plt.getOrCreateSubplot('RSI').addLine('Upper', level=70)

    # 3.7. Run Strategy
    meanStrategy2.run()

    # 3.8. Calculate Strategy Returns
    datesReturns = retAnalyzer.getReturns().getDateTimes()[:]
    dailyReturns = retAnalyzer.getReturns()[:]
    dailyReturns = pd.DataFrame(dailyReturns).set_index(pd.DatetimeIndex(datesReturns))

    # 3.9. Plot Strategy
    if chart == True:
        plt.plot(fromDateTime=dt.datetime(2016, 1, 1), toDateTime=dt.datetime(2016, 12, 31))

    return dailyReturns
```

```python
# 3.10. Plot Strategy Example
MeanStrategyRun2(14, 30, 70, True)
```

### Do Strategy Parameters Optimization and Calculate Performance Metrics

```python
# 4.1. Create Strategy Optimization Parameters Combinations
lowerThreshold = (30, 35)
upperThreshold = (65, 70)
pool = [lowerThreshold, upperThreshold]
```

```python
# 4.2. Calculate Benchmark Daily Returns
# 100% equity investment for buy and hold benchmark
data = pd.read_csv('./Advanced-Trading-Analysis-Data.txt', index_col='Date Time', parse_dates=True)
mean2DailyReturns = data['Adj Close'].pct_change(1)
mean2DailyReturns[0] = 0
mean2DailyReturns = pd.DataFrame(mean2DailyReturns)
```

```python
# 4.3. Do Strategy Optimization
mean2StartOptimization = time.time()
print('')
print('== Strategy Parameters Optimization ==')
print('')
print('Parameters Combinations (lowerThreshold, upperThreshold):')
for n in itertools.product(*pool):
    print(n)
    mean2DailyReturns.insert(len(mean2DailyReturns.columns), n, MeanStrategyRun2(14, n[0], n[1], False))
mean2EndOptimization = time.time()
mean2DailyReturns.columns = ['B&H', 'Mn2Ret1', 'Mn2Ret2', 'Mn2Ret3', 'Mn2Ret4']
print('')
print('Optimization Running Time: ', round(mean2EndOptimization - mean2StartOptimization, 4), ' seconds')
print('')
```

```python
# 4.4. Define Cumulative Returns and Performance Metrics Functions
def CumulativeReturns(dailyReturns):
    cumulativeReturns = np.cumprod(dailyReturns + 1) ** (252 / len(dailyReturns)) - 1
    return cumulativeReturns

def PerformanceMetrics(dailyReturns):
    annualizedReturn = (np.cumprod(dailyReturns + 1) ** (252 / len(dailyReturns)) - 1)[-1]
    annualizedStdDev = np.std(dailyReturns) * np.sqrt(252)
    annualizedSharpe = annualizedReturn / annualizedStdDev
    return annualizedReturn, annualizedStdDev, annualizedSharpe
```

```python
# 4.5. Chart Cumulative Returns Comparison
mean2CumulativeReturns = mean2DailyReturns.apply(CumulativeReturns, axis=0)
plt.plot(mean2CumulativeReturns['B&H'], label='B&H')
plt.plot(mean2CumulativeReturns['Mn2Ret1'], label='Mn2Ret1')
plt.plot(mean2CumulativeReturns['Mn2Ret2'], label='Mn2Ret2')
plt.plot(mean2CumulativeReturns['Mn2Ret3'], label='Mn2Ret3')
plt.plot(mean2CumulativeReturns['Mn2Ret4'], label='Mn2Ret4')
plt.title('Strategy Parameters Optimization Cumulative Returns')
plt.legend(loc='upper left')
plt.show()
```

```python
# 4.6. Calculate Performance Metrics and Print Summary Table
mean2PerformanceMetrics = mean2DailyReturns.apply(PerformanceMetrics, axis=0)
mean2PerformanceSummary = [{'0': 'Annualized:', '1': 'B&H', '2': 'Mn2Ret1', '3': 'Mn2Ret2', '4': 'Mn2Ret3',
                             '5': 'Mn2Ret4'},
        {'0': 'Return',
         '1': np.round(mean2PerformanceMetrics[0][0], 4), '2': np.round(mean2PerformanceMetrics[1][0], 4),
         '3': np.round(mean2PerformanceMetrics[2][0], 4), '4': np.round(mean2PerformanceMetrics[3][0], 4),
         '5': np.round(mean2PerformanceMetrics[4][0], 4)},
        {'0': 'Standard Deviation',
         '1': np.round(mean2PerformanceMetrics[0][1], 4), '2': np.round(mean2PerformanceMetrics[1][1], 4),
         '3': np.round(mean2PerformanceMetrics[2][1], 4), '4': np.round(mean2PerformanceMetrics[3][1], 4),
         '5': np.round(mean2PerformanceMetrics[4][1], 4)},
        {'0': 'Sharpe Ratio (Rf=0%)',
         '1': np.round(mean2PerformanceMetrics[0][2], 4), '2': np.round(mean2PerformanceMetrics[1][2], 4),
         '3': np.round(mean2PerformanceMetrics[2][2], 4), '4': np.round(mean2PerformanceMetrics[3][2], 4),
         '5': np.round(mean2PerformanceMetrics[4][2], 4)}]
mean2PerformanceSummary = pd.DataFrame(mean2PerformanceSummary)
print('')
print('== Strategy Parameters Optimization Performace Summary ==')
print('')
print(mean2PerformanceSummary)
print('')
```

### Do Multiple Hypothesis Testing P-Values Adjustments

```python
# 5.1. Calculate Multiple Hypothesis Testing P-Values
mean2MultipleTTests = mean2DailyReturns.iloc[:, 1:5].apply(st.ttest_1samp, axis=0, popmean=0, nan_policy='omit')
mean2MultiplePValues = (mean2MultipleTTests[0][1], mean2MultipleTTests[1][1], mean2MultipleTTests[2][1],
                         mean2MultipleTTests[3][1])
```

```python
# 5.2. Adjust Multiple Hypothesis Testing P-Values Calculations
mean2MultiplePValuesFWE = mt.multipletests(mean2MultiplePValues, alpha=0.05, method='bonferroni',
                                                   is_sorted=False, returnsorted=False)
mean2MultiplePValuesFDR = mt.multipletests(mean2MultiplePValues, alpha=0.05, method='fdr_bh',
                                                   is_sorted=False, returnsorted=False)
```

```python
# 5.3. Print Multiple Hypothesis Testing P-Values Adjustment Summary Table
mean2MultiplePValuesSummary = [{'0': '', '1': 'Mn2Ret1', '2': 'Mn2Ret2', '3': 'Mn2Ret3', '4': 'Mn2Ret4'},
        {'0': 'PValues',
         '1': np.round(mean2MultiplePValues[0], 4), '2': np.round(mean2MultiplePValues[1], 4),
         '3': np.round(mean2MultiplePValues[2], 4), '4': np.round(mean2MultiplePValues[3], 4)},
        {'0': 'PValues FWE',
         '1': np.round(mean2MultiplePValuesFWE[1][0], 4), '2': np.round(mean2MultiplePValuesFWE[1][1], 4),
         '3': np.round(mean2MultiplePValuesFWE[1][2], 4), '4': np.round(mean2MultiplePValuesFWE[1][3], 4)},
        {'0': 'PValues FDR',
         '1': np.round(mean2MultiplePValuesFDR[1][0], 4), '2': np.round(mean2MultiplePValuesFDR[1][1], 4),
         '3': np.round(mean2MultiplePValuesFDR[1][2], 4), '4': np.round(mean2MultiplePValuesFDR[1][3], 4)}]
mean2MultiplePValuesSummary = pd.DataFrame(mean2MultiplePValuesSummary)
print('')
print('== Multiple Hypothesis Testing P-Values Adjustments ==')
print('')
print(mean2MultiplePValuesSummary)
print('')
```

### Do Individual Time Series Bootstrap P-Value Multiple Comparison Adjustment

```python
# 6.1. Define Bootstrap Mean Function
def bmean(x):
    return x.mean(0)

# 6.2. Do Individual Time Series Bootstrap
mean2StartBoot = time.time()
print('')
print('== Individual Time Series Bootstrap ==')
print('')
mean2Boot = boot.CircularBlockBootstrap(block_size=10, x=mean2DailyReturns.iloc[:, 3])
mean2BootMeans = mean2Boot.apply(func=bmean, reps=1000)
mean2BootIntervals = mean2Boot.conf_int(func=bmean, reps=1000, method='percentile', size=0.95, tail='two')
mean2EndBoot = time.time()
print('')
print('Bootstrap Running Time: ', round(mean2EndBoot - mean2StartBoot, 4), ' seconds')
print('')
```

```python
# 6.3. Chart Individual Time Series Bootstrap Histogram
plt.hist(mean2BootMeans, bins=20, density=True, label='BootMeans')
plt.title('Population Mean Probability Distribution Simulation')
plt.axvline(x=np.mean(mean2DailyReturns.iloc[:, 3]), color='purple', linestyle='--', label='mean(Mn2Ret3)')
plt.axvline(x=np.mean(mean2BootMeans), color='red', linestyle='--', label='mean(BootMeans)')
plt.axvline(x=0, color='orange', linestyle='--')
plt.axvline(x=mean2BootIntervals[0], color='green', linestyle='--', label='BootLowerCI')
plt.axvline(x=mean2BootIntervals[1], color='green', linestyle='--', label='BootUpperCI')
plt.ylabel('Density')
plt.xlabel('Bin Edges')
plt.legend(loc='upper right')
plt.show()
```

```python
# 6.4. Calculate Individual Time Series Bootstrap P-Value
mean2BootPValue = 2 * min(np.mean(mean2BootMeans <= 0), np.mean(mean2BootMeans > 0))
```

```python
# 6.5. Adjust Individual Time Series Bootstrap P-Value Calculation
mean2BootPValueFWE = 1 - (1 - mean2BootPValue) ** 4
print('')
print('== Individual Time Series Bootstrap Hypothesis Testing ==')
print('')
print('Mn2Ret3 P-Value:', np.round(mean2BootPValue, 4))
print('Mn2Ret3 P-Value FWE:', np.round(mean2BootPValueFWE, 4))
```
